Import necessary libraries

```python
import numpy as np     # NumPy for numerical operations and array handling
import pandas as pd    # Pandas for data manipulation and analysis
import matplotlib.pyplot as plt  # Matplotlib for creating plots and charts
import seaborn as sns  # Seaborn for statistical data visualization
```

Load given dataset


```python
data = pd.read_csv("/content/india-state-wise-data-analysis.csv")   #dataset
data.head()   #read the dataset (first five entity)
```

Performing Exploratory Data Analysis (EDA)

```python
# Display basic information about the dataset
print(data.info())
```

Data Preporcessing

```python
# Check for missing values
print(data.isnull().sum())
```

```python
#spliting 'State and District' column
data[['district_code', 'state_name', 'district']] = data['State & District'].str.split(',', expand=True)
data.drop('State & District', axis=1, inplace=True)
```

```python
 # Display summary statistics
data.head()
```

```python
#split the info in column district code
for i in range(len(data)):
    data['district_code'][i] = data['district_code'][i].split(':')[-1]
```

```python
data[['Dist','District']]= data['district'].str.split(':', expand=True)
```

```python
#drop old column
data.drop(['district'],axis=1,inplace=True)
data.drop(['state'],axis=1,inplace=True)
```

```python
#data[['state','state_name']]= data['state_name'].str.split(':', expand=True)
```

```python
data.head()  #show database (first five entity)
```

Data visualisation

```python
# Create a bar plot for the population of each state
plt.figure(figsize=(14, 7))  # Create a new figure with specified size
plt.xlabel("States")         # Label for the x-axis
plt.ylabel("Population")     # Label for the y-axis
plt.bar(data['state_name'], data['Population']) # Create the bar chart
plt.xticks(rotation=90)      # Create the bar chart
plt.title("Population of Each State") # Set the title of the plot

 # Display the plot
plt.show()
```

```python
#plot the Literates distribution by Indian states
plt.figure(figsize=(14, 7))
plt.title('Literate Distribution by Indian State')
plt.xlabel('State')
plt.ylabel('Literate')
plt.bar(data["state_name"],data["Literate"])
plt.xticks(rotation=90)
# Show the plot
plt.show()
```

```python
# Create a bar plot for the Total education by each state
plt.figure(figsize=(14, 7))
plt.xlabel("States")
plt.ylabel('Total_Education')
plt.bar(data['state_name'], data['Total_Education'])
plt.xticks(rotation=90)
plt.title("Total_Education by State")
plt.show()
```

```python
# Create a bar plot for the Total education by each state
plt.figure(figsize=(14, 7))
plt.xlabel("States")
plt.ylabel('Total_Education')
plt.bar(data['state_name'], data['Total_Education'])
plt.xticks(rotation=90)
plt.title("Total_Education by State")
plt.show()
```

```python
#Group the data by state and calculate the total population, male, and female
grouped_data = data.groupby('state_name')[['Population', 'Male', 'Female']].sum()

# Create a stacked bar graph to visualize population, male, and female by state
plt.figure(figsize=(14, 7))
grouped_data.plot(kind='bar', stacked=True)

# Set plot title, axis labels, and rotation for x-axis labels
plt.title('Population, Male, and Female by Indian State')
plt.xlabel('State')
plt.ylabel('Value')
plt.xticks(rotation=90)

# Ensure proper layout and display the plot
plt.tight_layout()
plt.show()
```
